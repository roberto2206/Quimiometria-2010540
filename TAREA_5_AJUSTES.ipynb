```python
#PROBLEMA 1
import numpy as np
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [0.5, 2, 2.9, 3.5, 4]

def ln(x):
  return np.log(x)


def v(y):
  return y

n = len(ln(x))
m = (n * np.sum(ln(x) * v(y)) - (np.sum(ln(x)) * np.sum(v(y)))) / (n * np.sum(ln(x)**2) - (np.sum(ln(x)))**2)
c = (np.sum(v(y)) - m * np.sum(ln(x)))/n

regresion = m*ln(x) + c

print(f'Ecuacion de regresion: y = {m:.2f}x + {b:.2f}')
print("Sabiendo que la ecuacion transformada tiene la forma de y = a*ln(x) + b,","\npodemos compararlo con la linea de tendencia obtenida, concluyendo asi que:","\na = ", m, "\nb =", c)

x1 = float(input("Introduzca el Valor de x que quiera extrapolar / interpolar: "))
evaluado = m*ln(x1) + c

print("El valor de x =",x1, "corresponde a un valor de y =",evaluado)

plt.scatter(ln(x), v(y))
plt.plot(ln(x), regresion, color='purple')
plt.xlabel('ln(x)')
plt.ylabel('y')
plt.title('Regresión por mínimos cuadrados')
plt.show()


```

```python
#PROBLEMA 3
import numpy as np
import matplotlib.pyplot as plt

T = np.array([-50, -30, 0, 60, 90, 110])
c = np.array([1270, 1280, 1350, 1480, 1580, 1700])

n = len(T)
m = (n * np.sum(c * T) - (np.sum(T) * np.sum(c))) / (n * np.sum(T**2) - (np.sum(T))**2)
b = (np.sum(c) - m * np.sum(T))/n

regresion = m*T + b

print(f'Ecuacion de regresion: c(T) = {m:.3f}T + {b:.3f}')

plt.scatter(T, c)
plt.plot(T, regresion, color='purple')
plt.xlabel('T')
plt.ylabel('c(T)')
plt.title('Regresión por mínimos cuadrados')
plt.show()
```

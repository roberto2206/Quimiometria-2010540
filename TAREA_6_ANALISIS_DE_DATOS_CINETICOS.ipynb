```python
#PROBLEMA 2
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from sympy import integrate,evalf
from sympy.abc import T
from sympy.abc import t



f1= 1/(T - 70)
f2 = 1

I1 = integrate(f1, (T,300,200)).evalf()
I2 = integrate(f2,(t,0,3)).evalf()



def ecuacion (T, t, k):
   dTdt = k * (T-Tm)
   return dTdt


Tm = 70
T0 = 300
k = float(I1/I2)
t = np.linspace(0, 100, 100)


Sol = odeint(ecuacion, T0, t, args=(k,))


plt.plot(t, Sol, color ='purple')
plt.xticks(np.arange(0,100,20))
plt.yticks(np.arange(60,300,10))
plt.axhline(y=70, xmin=0, xmax=1, color='r', linestyle= '--')
plt.title("Método Gráfico")
plt.xticks(np.arange(0,100,10))
plt.yticks(np.arange(60,300,10))
plt.grid(True)
plt.title('Grafica')
plt.xlabel('Tiempo')
plt.ylabel('Temperatura')
plt.legend()
plt.show()
```

```python
#PROBLEMA 3A
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt


def sistema(t, y):
  dxdt =  1* y[0] - 0.1 * y[0]**2 - 0.05 * y[0] * y[1]
  dydt = 1.7 * y[1] - 0.1 * y[1]**2 - 0.15 * y[0] * y[1]
  return [dxdt, dydt]

con_in = [1.1, 1]
time_span = (0, 2000)

solucion = solve_ivp(sistema, time_span, con_in,  dense_output=True)


t_fine = np.linspace(0, 500, 1000)
y_fine = solucion.sol(t_fine)


plt.figure(figsize=(10, 6))
plt.plot(t_fine, y_fine[0], color = 'purple')
plt.plot(t_fine, y_fine[1], color = 'b')
plt.title('Competencia Poblaciones')
plt.xlabel('t')
plt.ylabel('Poblacion')
plt.legend()
plt.grid(True)
plt.show()
```
